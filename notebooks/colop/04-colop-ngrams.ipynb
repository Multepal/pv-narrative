# Colop Ngrams

```python
import pandas as pd
import numpy as np
import nltk
from nltk.util import ngrams
# import spacy
```

```python
DOC = pd.read_csv("colop-DOC.csv").set_index(['parte','capit'])
```

```python
SENT = DOC.doc_str.apply(nltk.sent_tokenize).apply(pd.Series).stack().to_frame('sent_str')
SENT.index.names = ['parte','capit','sent']
```

```python
SENT
```

```python
BG = SENT.sent_str.str.replace(r"[^A-Za-z' ]", "", regex=True).str.replace("-","")\
    .str.split()\
    .apply(lambda x: list(ngrams(x, 2)))\
    .apply(pd.Series).stack()\
    .apply(pd.Series)
```

```python
BG_DOC = TG.groupby(['parte','capit']).value_counts().to_frame('n')
```

```python
BG_DOC[BG_DOC.n > 2]
```

```python
BG_COUNT = TG.value_counts().to_frame('n')
```

```python
BG_COUNT[BG_COUNT.n > 3].sort_index().tail(30)
```
