# Colop Glossary

```python
import pandas as pd
import numpy as np
```

```python
GLOSS = pd.read_csv("colop-glossary.txt", sep="|").set_index('term')
```

```python
GLOSS
```

```python
DOC = pd.read_csv("colop-DOC.csv").set_index(['parte','capit'])
```

# CCC

```python
matches = []
for doc in DOC.iterrows():
    doc_id = doc[0]
    doc_str = doc[1].doc_str.lower()
    for gloss in GLOSS.iterrows():
        gloss_str = gloss[0].lower()
        if gloss_str in doc_str:
            matches.append((doc_id[0], doc_id[1], gloss[0]))
CCC = pd.DataFrame(matches, columns=['parte','capit','term'])
CCC.index.name = 'ccc_id'
```

```python
CCC
```

```python
PAIR = CCC.merge(CCC, on=['parte','capit'])\
    .value_counts(['term_x','term_y'])\
    .unstack(fill_value=0).corr('kendall')
```

```python
PAIR['SUYWA'].sort_values(ascending=False).head(20)
```

```python
PAIR2 = PAIR.stack().to_frame('r')
```

```python
PAIR2.index.names = ['x','y']
```

```python
PAIR3 = PAIR2.query("x < y").sort_values('r', ascending=False)
```

```python
PAIR3[PAIR3.r > .8]
```

# Save

```python
for df in [(GLOSS, 'GLOSS'), (PAIR,'CCCPAIR')]:
    df[0].to_csv(f"colop-{df[1]}.csv", index=True)
```
