# Edmonson Chapter Titles

```python
import pandas as pd
import numpy as np
import roman
```

```python
toc = open("edmonson-toc.md", "r").readlines()
```

```python
data = []
part_nums = list("ABCD")
for line in toc:
    parts = line.split()
    
    if len(parts) == 5 and parts[3] == 'CREATION':
        part_num = part_nums.index(parts[0][:-1]) + 1
        continue
    
    chap_num = roman.fromRoman(parts[0][:-1])
    chap_line = parts[-1]
    chap_title = ' '.join(parts[1:-1])

    data.append((part_num, chap_num, chap_title, chap_line))
```

```python
TOC = pd.DataFrame(data, columns=['part_num','chap_num','chap_title','chap_line'])
TOC = TOC.set_index(['chap_num'])
```
