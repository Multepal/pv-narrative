# Christenson Extract Lines from Literal

Actual number of lines = $8726$.

Christenson misnumbers lines:
- $4910$ is at $4908$.
- $5260$ repeated twice, throwing off the rest.

**NOW FIXED.**

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
lines = pd.DataFrame(open("christenson-pv-literal.txt", "r").readlines(), columns=['line_str'])
lines = lines[~lines.line_str.str.match(r"^#")]
lines.line_str = lines.line_str.str.strip()
page_num_pat = r"\s*\d+\s*$"
lines['page_num'] = lines[lines.line_str.str.match(page_num_pat)].line_str
lines.page_num = lines.page_num.bfill()
lines = lines[~lines.line_str.str.match(page_num_pat)].copy()
lines = lines.dropna()
lines = lines[~lines.line_str.str.match(r"^\s*$")]
lines = lines.reset_index(drop=True)
lines.index = lines.index + 1
lines.index.name = 'line_num'
```

```python
LINE = lines.line_str.str.split(r"\s\s+").apply(pd.Series).rename(columns={0:'eng_str', 1:'quc_str', 2:'line_marker'})
LINE.line_marker = LINE.line_marker.fillna('')
LINE.eng_str = LINE.eng_str.str.replace(r"\d+", "", regex=True)
LINE.quc_str = LINE.quc_str.str.replace(r"\d+", "", regex=True)
```
