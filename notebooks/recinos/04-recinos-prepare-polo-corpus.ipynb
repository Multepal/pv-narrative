```python
import pandas as pd

POLO = pd.read_csv('recinos-DOC.csv')
POLO = POLO.rename(columns={'label':'doc_label'})
POLO = POLO.rename(columns={'parte':'doc_part'})
POLO = POLO.rename(columns={'parte':'doc_part'})
POLO = POLO.rename(columns={'doc_str':'doc_content'})
POLO['doc_key'] = POLO.apply(lambda x: f"{str(x.doc_part).zfill(2)}:{str(x.capit).zfill(2)}", axis=1)
POLO['doc_title'] = POLO['doc_key'] + ' ' + POLO['doc_content'].str[:40]
POLO[['doc_key', 'doc_label','doc_part','doc_title', 'doc_content']].to_csv('corpus.csv', sep='|', index=False)
```
