# Recinos Noun Phrases

```python
import pandas as pd
from unidecode import unidecode
```

```python
SPACY = pd.read_pickle("spacy.pickle")
```

```python
NP = SPACY.apply(lambda x: ["".join(str(np)) for np in x.noun_chunks]).apply(pd.Series).stack().to_frame('np')
```

```python
NP['np'] = NP.np.apply(unidecode).str.upper().str.replace("\W+", "", regex=True).str.strip()
```

```python
NP.sort_values('np')
```

```python
NPM = NP.groupby(['parte', 'capit', 'np']).np.count().unstack(fill_value=0)
```
