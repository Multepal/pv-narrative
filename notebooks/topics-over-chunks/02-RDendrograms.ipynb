```r
library(tidyverse)  # data manipulation
library(cluster)    # clustering algorithms
library(factoextra) # clustering visualization
library(dendextend) # for comparing two dendrograms
```

```r
df <- read.csv("output//COLQ_TFIDF_SIM.csv")
```

```r
df <- scale(df)
```

```r
d <- dist(df, method = "euclidean")
```

```r
hc1 <- hclust(d, method = "ward.D2")
hc2 <- agnes(df, method = "ward")
hc3 <- diana(df)
```

```r
# sub_grp <- cutree(hc1, k = 6)
```

```r
plot(hc1, cex = 0.6, hang = -1, main="Dendrogram of TFIDF SIM")
rect.hclust(hc1, k = 6, border = 2:5)
```

```r
pltree(hc2, hang=-1, cex=.6, main="Dendrogram of TFIDF")
```

```r
pltree(hc3, hang=-1, cex=.6, main="Dendrogram of TFIDF")
```
